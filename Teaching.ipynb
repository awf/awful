```python
#@title

# Maths functions
import numpy as np
from numpy import sin, cos

number_range = np.arange

# Functions from https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html
import matplotlib.pyplot as plt

def plot(*args, **kwargs):
    plt.plot(*args, *kwargs)
    plt.gca().set_aspect('equal')

arrow = lambda *args, **kwargs: plt.arrow(*args, width=0.01, **kwargs)
line = lambda *args, **kwargs: plt.arrow(*args, head_length=0, head_width=0, **kwargs)

```

# Vectors and plotting

```python
# Basics: points, lines, arrows
plot(1.2,3.4, 'k.')
plot(2.2,3.3, 'b.')
plot(1.2,4.3, 'r.')

arrow(1.2,4.3, 0, -.6)
```

```python
# No need for wolfram alpha - just do it !
for x in number_range(0, 6.28, 0.1):
  plot(x, sin(x), '.')
```

```python
for x in number_range(0, 6.28, 0.1):
  plot(cos(x), sin(x), '.')
```

# Let's compute an integral

```python
for x in number_range(0, 3.14, 0.1):
  plot(x,sin(x), '.')
  arrow(x, 0, 0, sin(x))

total = 0
for x in number_range(0, 3.14, 0.1):
  total = total + sin(x)
print('Total  = ', total)
print('Answer = ', -cos(3.14) - -cos(0))
```

```python
for x in number_range(0, 3.14, 0.1):
  plot(x,sin(x), '.')
  line(x, 0, 0, sin(x), width=0.1)

total = 0
for x in number_range(0, 3.14, 0.1):
  total = total + sin(x) * 0.1
print('Total  = ', total)
print('Answer = ', -cos(3.14) - -cos(0))
```
